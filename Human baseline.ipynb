```python
import pandas as pd
import re
from scipy.stats import kendalltau
from statistics import mean
```

```python
answers = pd.read_excel('D:\\Диплом_текстовые_квесты\\Текстовые квесты (Ответы).xlsx')
```

```python
correct_orders = {
    1:['А','В','Б','Г'],
    2:['И','К','А','Б','З','В','Д','Г','Ж','Е'],
    3:['В','А','Г','Б','Ж','Д','Е'],
    4:['Д','Ж','А','И','Б','Г','Е','В','З'],
}
```

```python
def get_user_orders(ordering_task_idx, answers_df):
    current_task_columns = [column_name for column_name in answers_df.keys() if column_name.startswith(str(ordering_task_idx)) and '[' in column_name]
    column2option = {}
    for column_name in current_task_columns:
        if '[' in column_name:
            column2option[column_name] = column_name.split('[')[1][0]
    sequences = []
    for i, row in answers_df.iterrows():
        sequence = [None for i in range(len(column2option))]
        for column_name, option in column2option.items():
            position = row[column_name]-1
            sequence[position] = option
        sequences.append(sequence)
    return sequences
```

```python
ordering_answers = {}
for i in range(1, 5):
    ordering_answers[i] = get_user_orders(i, answers)
```

```python
ordering_answers
```

```python
def longest_correct_subsequence(predicted, correct):
    correct_transitions = set([(item_from, item_to) for item_from, item_to in zip(correct[:-1], correct[1:])])
    predicted_transitions = [(item_from, item_to) for item_from, item_to in zip(predicted[:-1], predicted[1:])]
    predicted_transitions_are_correct = [transition in correct_transitions for transition in predicted_transitions]
    
    longest_correct_transitions_subsequence = 0
    current_correct_transitions_subsequence = 0
    #Add False as last item to include last real item checking into the loop
    for predicted_transition_is_correct in predicted_transitions_are_correct + [False]:
        if predicted_transition_is_correct:
            current_correct_transitions_subsequence += 1
        else:
            if current_correct_transitions_subsequence > longest_correct_transitions_subsequence:
                longest_correct_transitions_subsequence = current_correct_transitions_subsequence
            current_correct_transitions_subsequence = 0
    #return number of items in longest correct sequence (not number of transitions)
    return longest_correct_transitions_subsequence + 1 if longest_correct_transitions_subsequence else 0  
```

```python
human_ordering_correlation = pd.DataFrame(columns=['length', "mean Kendall's tau", "mean longest subsequence"])
for sample_idx, correct_order in correct_orders.items():
    predicted_orders = ordering_answers[sample_idx]
    mean_tau = mean([kendalltau(predicted_order, correct_order).correlation for predicted_order in predicted_orders])
    mean_longest_subsequence = mean([longest_correct_subsequence(predicted_order, correct_order) for predicted_order in predicted_orders])
    human_ordering_correlation.loc[len(human_ordering_correlation)] = len(correct_order), mean_tau, mean_longest_subsequence
```
